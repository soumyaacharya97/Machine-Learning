**Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

**A-**

Machine learning is a artificial inteligence which learns pattern from data and make decision from it.

#### 1. Supervised Learning :

The algorithm learns from labeled training data with known outcomes.

#### How it works:
It maps inputs to the correct outputs like a student learning with an answer key.

Example: Email spam filters analyze past emails marked as "spam" or "not spam" to classify new incoming messages.

#### 2. Unsupervised Learning:

The algorithm finds hidden patterns or structures in unlabeled data.

#### How it works: It groups data based on inherent similarities without any guiding output labels.
Example: E-commerce sites group customers into segments based on similar purchasing habits for targeted marketing.

**Q2. What is Supervised Learning? Explain the difference between Regression and Classification.**

**A-**

The algorithm learns from labeled training data with known outcomes.

### Regression :

Regression is about "How Much?"

#### The Concept: You are predicting a point on an endless number line. The output can be any fraction, decimal, or whole number.
#### Analogy: Guessing the exact weight of a suitcase by looking at it (e.g., 14.3 kg, 22.1 kg).
#### Visual: A smooth, continuous line cutting through a scatter plot of data points to find a trend.

### Classification :

Classification is about "Which One?"

#### The Concept: You are sorting items into separate, pre-defined buckets. The output must belong to a specific group.
#### Analogy: Sorting luggage at the airport into "Carry-on" or "Checked" bins.
####Visual: A sharp boundary line or wall drawn on a graph to separate data points into distinct clusters.

**Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

**A-**

Linear regression is a supervised learning algorithm used to predict a continuous numerical output by finding the best-fit straight line through the data points.

The equation of Linear regression is :                               

y=mx+c

y is the Dependent Variable (the target prediction).

m is the Slope, which maps directly to the Coefficient. It controls the steepness of the line.

x is the Independent Variable (the input feature).

c is the Y-intercept, which maps directly to the Intercept. It is where the line hits the vertical axis.

**Q4. Explain the following evaluation metrics used in Linear Regression:**

**SSE**

**MSE**

**RMSE**

**R² Score**

**Adjusted R²**

**A-**

### 1. SSE (Sum of Squared Errors) :
SSE calculates the total squared distance between the actual data points and the model's predicted values. It represents the raw volume of error that the model fails to explain.

### 2. MSE (Mean Squared Error) :
MSE solves the dataset size problem by finding the average of the squared errors. It measures the variance of the residuals.

### 3. RMSE (Root Mean Squared Error) :
RMSE is simply the square root of the Mean Squared Error. It functions as the standard deviation of the model's residuals.

### 4. (R^2) Score (Coefficient of Determination) :
Unlike the previous metrics that measure absolute error, (R^2) is a relative metric that quantifies how much of the variance in the target variable (y) can be explained by your independent variables (x).

#### 5. Adjusted (R^2) :
Adjusted (R^2) is a modified version of the (R^2) score specifically engineered to fix the flaw mentioned above. It accounts for the number of predictors (features) in the model.


**Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.**

**A-**

### 1. Data Collection :

This initial step involves gathering relevant data from various sources to solve the defined business problem.

### 2. Data Cleaning & Preprocessing :

Raw data is almost always messy and unsuited for immediate model training. This step refines that data.

### 3. Feature Engineering & Selection :

This step involves transforming raw variables into meaningful inputs that help machine learning algorithms learn more effectively.

### 4. Data Splitting :

To evaluate a model fairly, you cannot test it on the same data it learned from. Data must be split into independent sets.

### 5. Model Training :

This is the step where the machine learning algorithm actually analyzes the training data to discover patterns and learn the mathematical relationships.

### 6. Model Evaluation :

The final step measures how well the trained model performs on unseen, real-world data to ensure it can generalize well instead of just memorizing the training set

**Q9. Split the dataset into training and testing sets (80% training, 20% testing).**

```python
from sklearn.model_selection import train_test_split

X = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio']]
y = df[['b', 'lstat', 'medv']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
## Make predictions

from sklearn import metrics

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on test data
y_pred = model.predict(X_test)
```

```python
## Calculate MAE

mae = metrics.mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (sklearn): {mae:.4f}")
```

```python
## Calculate MSE

mse = metrics.mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (sklearn): {mse:.4f}")
```

```python
## Calculate RMSE

mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (sklearn): {rmse:.4f}")
```

```python
## Calculate R² Score

r2 = metrics.r2_score(y_test, y_pred)
print(f"R² Score (sklearn): {r2:.4f}")
```
